```python
import io
import json
import pandas as pd
import requests
import time

API_KEY = "824d4b3157be6c75e79982be9f7ee822770f37d5aa6542fcb260ce492375"
```

```python
r = requests.get(
        "https://data-api.defipulse.com/api/v1/defipulse/api/getLendingHistory?", 
        params={
            'api-key': API_KEY
        }
)
print(r.status_code)
df = pd.DataFrame(json.loads(r.content.decode()))
df.timestamp = pd.to_numeric(df.timestamp, errors='coerce')
df = df.dropna()
df['date'] = pd.to_datetime(df.timestamp, unit='s').dt.date

def get_usdt(r, which="USDT"):
    try:
        return r[which]['USD']
    except:
        return 0

df['usdt'] = df.outstanding.apply(get_usdt, which="USDT")
df['total'] = df.outstanding.apply(get_usdt, which="total")
df.tail()
```

```python
df.to_csv('lending.csv', index=False)
```
